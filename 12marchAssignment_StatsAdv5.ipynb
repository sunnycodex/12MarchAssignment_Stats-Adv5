```python
#1
import scipy.stats as st

# Sample mean and standard deviation
mean = 50
std_dev = 5
sample_size=70
n=17

# Confidence level
alpha = 0.95

# Calculate the standard error of the mean
sem = std_dev / (n ** 0.5)

# Calculate the confidence interval
conf_int = st.norm.interval(alpha, loc=mean, scale=sem)

print(f"The 95% confidence interval is: {conf_int}")
```

```python
#6
import numpy as np
import matplotlib.pyplot as plt
from scipy.stats import chi2

# Degrees of freedom
df = 10

# Values
x = np.linspace(chi2.ppf(0.01, df), chi2.ppf(0.99, df), 100)
rv = chi2(df)

# Chi-square statistic
chi_stat = 15

# Plot the chi-square distribution
plt.plot(x, rv.pdf(x), 'k-', lw=2, label='chi2 pdf')

# Shade the area under the curve corresponding to chi_stat
x_fill = np.linspace(chi_stat, chi2.ppf(0.99, df), 100)
plt.fill_between(x_fill, rv.pdf(x_fill), color='red', alpha=0.5)

# Label the axes
plt.xlabel('Value')
plt.ylabel('Frequency')
plt.title('Chi-Square Distribution with 10 Degrees of Freedom')

# Show the plot
plt.show()
```

```python
#7
import math

sample_size = 1000
coke_preferred = 520
confidence_level = 0.99
z_score = 2.576  # Z-score for a 99% confidence level

sample_proportion = coke_preferred / sample_size

standard_error = math.sqrt((sample_proportion * (1 - sample_proportion)) / sample_size)

margin_of_error = z_score * standard_error

# Calculate the confidence interval
lower_bound = sample_proportion - margin_of_error
upper_bound = sample_proportion + margin_of_error

print(f"99% Confidence Interval: ({lower_bound:.4f}, {upper_bound:.4f})")
```

```python
#9
import scipy.stats as stats

# Observed frequencies
observed = [[60, 140], [30, 170]]
sig_val=0.09

# Conduct the chi-square test for independence
chi2_stat, p_val, dof, expected = stats.chi2_contingency(observed)

print(f"Chi-square statistic = {chi2_stat}")
print(f"P-value = {p_val}")
print(f"Degrees of freedom = {dof}")
print(f"Expected frequencies = {expected}")


if p_val<sig_val:
    print("Reject the null hypothsis")
else:
    print("failed to reject the null hypothesis")
```

```python
#11
import scipy.stats as stats
import numpy as np
siglev=0.05
# Sample mean
mean = 72
# Population mean (under null hypothesis)
mu = 70
# Sample standard deviation
std_dev = 10
# Sample size
n = 30

# Calculate the standard error
std_err = std_dev / (n ** 0.5)

# Calculate the t statistic
t_stat = (mean - mu) / std_err

# Degrees of freedom
df = n - 1

# Calculate the p-value (two-tailed test)
p_val = stats.t.sf(np.abs(t_stat), df)*2

print(f"T statistic = {t_stat}")
print(f"P-value = {p_val}")

if p_val<siglev:
    print("Reject the null hypotheis")
else:
    print("Accept the null hypothesis")
    
print("there is no evidence to suggest that the population mean is significantly different from 70")
```
